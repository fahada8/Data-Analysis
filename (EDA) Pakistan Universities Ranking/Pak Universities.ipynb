# Data Wrangling

```python
data1[['word_ranking','World_total']] = data1['word_ranking'].str.split('/', expand = True) 
```

```python
data1[['asia_ranking','asia_total']] = data1['asia_ranking'].str.split('/', expand = True)
```

```python
data1[['pakistani_ranking','asia_total']] = data1['pakistani_ranking'].str.split('/', expand = True)
```

### Dropping Irrelevant Columns from Dataset

```python
data1.drop(columns=['World_total','asia_total','web','language','phone','address'], axis=1, inplace=True)
```

```python
data1.sample(10)
```

### Renaming Columns 

```python
data1.rename(columns = {'name':'University_Name','word_ranking':'World_Ranking','asia_ranking':'Asia_Ranking',
                     'pakistani_ranking':'HEC_Ranking','city':'Location','type':'Type','acceptance':'Acceptance_Rate',
                     'publication':'Publications'}, inplace = True)
```

```python
data1.head(10)
```

```python
data1['Publications'] = data1['Publications'].replace({',':''},regex=True)
```

```python
data1[['Type','X']] = data1.Type.str.split('\n', expand=True)
```

```python
data1['Acceptance_Rate'] = data1['Acceptance_Rate'].replace({'%':''},regex=True)
```

```python
data1['Acceptance_Rate'] = data1['Acceptance_Rate'].replace({',':''},regex=True)
```

```python
data1.drop(columns=['X'],inplace=True)
```

```python
data1.head(10)
```

### Handling Missing (NAN) values 

```python
# Removing NAN Values from Acceptance Rate
data1['Acceptance_Rate'] = data1['Acceptance_Rate'].astype(str).astype(float)
data1['Acceptance_Rate'] = data1['Acceptance_Rate'].replace(np.nan,data1['Acceptance_Rate'].mean())
```

```python
# Removing NAN Values from Publications
data1['Publications'] = np.nan_to_num(data1['Publications']).astype(float)
data1['Publications'].replace(np.nan, data1['Publications'].mean() , inplace=True)
```

```python
# Type Casting
data1['World_Ranking'] = data1['World_Ranking'].astype(int)
data1['Asia_Ranking'] = data1['Asia_Ranking'].astype(int)
data1['HEC_Ranking'] = data1['HEC_Ranking'].astype(int)
data1['Location'] = data1['Location'].astype('category')
data1['Type'] = data1['Type'].astype('category')
data1['Acceptance_Rate'] = data1['Acceptance_Rate'].astype(int)
data1['Publications'] = data1['Publications'].astype(int)
```

### <b>ANALYSIS REPORT : </b> 

Based on the data analysis performed, the following conclusions have been reached:

<b> (1)</b>  The University of Punjab is ranked as the top university in Pakistan according to World, Asian, and Higher Education Commission (HEC) rankings.

<b> (2)</b> The National University of Sciences and Technology (NUST) has the lowest acceptance rate among the universities in Pakistan. This may indicate that NUST is a highly competitive institution with high admission standards.

<b> (3)</b> The University of Management and Technology (UMT) has the highest acceptance rate among the universities in Pakistan. This may suggest that UMT is more accessible and less selective compared to other universities in the country.

<b> (4)</b> The University of Agriculture, Faisalabad (UAF) has the highest number of publications among the universities in Pakistan. This may indicate that UAF is a highly productive and research-oriented institution.

<b> (5)</b> On the other hand, Beaconhouse National University (BNU) has the lowest number of publications among the universities in Pakistan. This may suggest that BNU is less focused on research compared to other universities in the country.

<b> (6) </b> It can also be noted that most of the institutions in the dataset are located in major cities in Pakistan including Lahore, Islamabad, Karachi, and Peshawar. This highlights the concentration of higher education institutions in these urban areas. Furthermore, the majority of the universities in the dataset are public institutions, which are funded by the government and offer education to students at a relatively low cost. This indicates that public universities play a significant role in providing access to higher education in Pakistan.

<b> These conclusions are based on the data analyzed, and it is important to note that other factors such as the quality of education and student experiences may also play a role in determining the overall reputation of a university. </b>
